## Logistic Growth

```python
import numpy as np
import matplotlib.pyplot as plt

r = 0.1
k = 30e5
h = 0.5
t = np.arange(0,10,0.5)

def logistic_model (t,N):
    return r*N*(1-N/k)

N = np.zeros(len(t))
N[0] = 3e5

for i in range(0, len(t) -1):
    N[i+1] = N[i] + h*logistic_model(t[i],N[i])

plt.plot(t, N)
plt.title('logistic model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

## Predation Model

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,100,0.5)
N = np.zeros(len(t))

r = 0.1
# k = 5
k = 30
h = 0.5
A = 5
B = .2
N[0] = 3

def predation_model (t,N):
    return r*N*(1-N/k) - (B*N*N)/(A*A+N*N)

for i in range(0, len(t) -1):
    N[i+1] = N[i] + h*predation_model(t[i],N[i])

plt.plot(t, N)
plt.title('Predation Model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

## Prey-Predator

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,100,0.5)
N = np.zeros(len(t))
P = np.zeros(len(t))

k = 100
h = 0.5
alpha_1 = 0.1
alpha_2 = 0.2
beta = 1
gamma = 0.5

N[0] = 2
P[0] = .5

def NNN (t, N, P):
    return alpha_1*N*(1 - N/k - beta*P)

def PP (t, N, P):
    return (-1)*alpha_2*P*(1 - gamma*N)

for i in range(0, len(t) -1):
    N[i+1] = N[i] + h * NNN(t[i], N[i], P[i])
    P[i+1] = P[i] + h * PP(t[i], N[i], P[i])

plt.plot(t, N, color='#c356fd')
plt.plot(t, P, color='#00e5f7')
plt.legend(["Prey","Predator"],loc="lower right")
plt.title('Prey')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

## NEW Prey-Predator

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,100,0.05)
N = np.zeros(len(t))
P = np.zeros(len(t))

k = 100
h = 0.05
alpha = 0.1
beta = 1

N[0] = 5
P[0] = 2

def NNN (t, N, P):
    return alpha*N - beta*P - N*(N*N + P*P)

def PP (t, N, P):
    return beta*N + alpha*P - P*(N*N + P*P)

for i in range(0, len(t) -1):
    N[i+1] = N[i] + h * NNN(t[i], N[i], P[i])
    P[i+1] = P[i] + h * PP(t[i], N[i], P[i])

plt.plot(t, N, color='#FF00FF')
plt.plot(t, P, color='#0000FF')
plt.legend(["Prey","Predator"],loc="lower right")
plt.title('Prey')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

## Bacteria thingy

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,100,0.05)
N = np.zeros(len(t))
P = np.zeros(len(t))

k = 100
h = 0.05
alpha = 0.1
beta = 0.2
delta = 0.1

N[0] = 5
P[0] = 2

def NNN (t, N, P):
    return alpha*N**(2/3) - delta*N

def PP (t, N, P):
    return beta*P - r*N*N

for i in range(0, len(t) -1):
    N[i+1] = N[i] + h * NNN(t[i], N[i], P[i])
    P[i+1] = P[i] + h * PP(t[i], N[i], P[i])

plt.plot(t, N, color='#FF00FF')
plt.plot(t, P, color='#0000FF')
plt.legend(["Prey","Predator"],loc="lower right")
plt.title('Prey')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```
