```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.001
n0 = 3*10**5
lam = 0.05
t = 100

def f(lam, n):
    return lam*n

i = 0
n = n0
n_val = []
time = []
while(i < t):
    n += h*f(lam, n)
    i += h
    n_val.append(n)
    time.append(i)

plt.plot(time, n_val)
# plt.axes().set_ylim(bottom=0)
plt.xlabel('time')
plt.ylabel('population')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.001
n0 = 3*(10**5)
t = 10
k = 2*(10**5)
r = 0.5

def f(k, r, n):
  return r*n*(1-(n/k))

i = 0
n = n0
n_val = []
time = []
while(i < t):
    n += h*f(k, r, n)
    i += h
    n_val.append(n)
    time.append(i)

plt.plot(time, n_val)
# plt.axes().set_ylim(bottom=0)
# plt.axes().set_xlim(left=0)
plt.xlabel('time')
plt.ylabel('population')
plt.show()
```
