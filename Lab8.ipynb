```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

#### Sub-Problem-1

```python
t_min = 0
t_max = 10
x_min = 0
x_max = 10

c = 1

h_x = (int)((x_max)/0.01+2)
h_t = (int)((t_max)/0.01+1)

dx = 0.01
dt = 0.01

F = 0.05

X = np.linspace(0, x_max, h_x)
T = np.linspace(0, t_max, h_t)

u = np.zeros((np.shape(X)[0], np.shape(T)[0]))

for i in range(np.shape(X)[0]):
  u[i][0] = 0.25*math.sin(i*dx)

for i in range(np.shape(T)[0]):
  u[np.shape(X)[0]-1][i] = 0
```

```python
for n in range(1, np.shape(T)[0]):
    for i in range(np.shape(X)[0]-1):
        u[i,n] = u[i,n-1]+((u[i+1,n-1]-u[i,n-1])*dt)/(c*dx)
```

```python
plt.plot(T[0:30], u[1, 0:30], color='purple')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity Profile')
plt.show()
```

```python
plt.plot(X,u[:, 0], color='red')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Velocity Profile')
plt.show()
```

```python
plt.plot(X, u[: , 30], color='purple')
plt.xlabel('Displacement')
plt.ylabel('Velocity')
plt.title('Velocity Profile')
plt.show()
```

#### Sub-Problem-2

```python
t_min = 0
t_max = 10
x_min = 0
x_max = 1

c = 1

h_x = (int)((x_max)/0.01+2)
h_t = (int)((t_max)/0.01+1)

dx = 0.01
dt = 0.01

F = 0.05

X = np.linspace(0, x_max, h_x)
T = np.linspace(0, t_max, h_t)

u = np.zeros((np.shape(X)[0], np.shape(T)[0]))

for i in range(np.shape(X)[0]):
  u[i][0] = 0.25*math.sin(i)

for i in range(np.shape(T)[0]):
  u[np.shape(X)[0]-1][i] = 0
```

```python
for n in range(1, np.shape(T)[0]):
    for i in range(np.shape(X)[0]-1):
        u[i,n] = u[i,n-1]+((u[i+1,n-1]-2*u[i,n-1]-u[i-1,n-1])*dt)/(c*dx*dx)
```

```python
plt.plot(T, u[30, :])
plt.xlabel('t')
plt.ylabel('u')
plt.title('Velocity Profile')
plt.show()
```

```python
plt.plot(X, u[: , 30], color='purple')
plt.xlabel('Displacement')
plt.ylabel('Velocity')
plt.title('Velocity Profile')
plt.show()
```
