```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

## Q1


### PGM Single Species

```python
# code
r = 0.5
h = 0.05

t = np.arange(0,10,h)
N = np.zeros(len(t))

U = 200
K = 450
N[0] = 350

def PGMSS(t, N):
    return r * N * (N/U - 1) * (1 - N/K)

for i in range(0, len(t) -1):
    N[i+1] = N[i] + PGMSS(t[i], N[i]) * h

plt.plot(t, N, color='#FF00FF')
plt.title('Population Model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

## Q2

### PGM with Prey-Predator Interaction

```python
# code
h = 0.05
alpha1 = 0.489
alpha2 = 0.982
beta1 = 2
beta2 = 0.8

t = np.arange(0,50,h)
N = np.zeros(len(t))
P = np.zeros(len(t))

N[0] = 3
P[0] = .8

def NNN (t, N, P):
    return alpha1*N - beta1*N*P

def PP (t, N, P):
    return -1*alpha2*P + beta2*N*P

for i in range(0, len(t) -1):
    N[i+1] = N[i] + h * NNN(t[i], N[i], P[i])
    P[i+1] = P[i] + h * PP(t[i], N[i], P[i])

plt.plot(t, N, color='#c356fd')
plt.plot(t, P, color='#00e5f7')
plt.legend(["Prey","Predator"],loc="lower right")
plt.title('Prey - Predator')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

#### Phase Diagram

```python
# phase plot
plt.plot(N, P, color='#c356fd')
plt.title('Phase Diagram')
plt.xlabel('Prey')
plt.ylabel('Predator')
plt.show
```

## Q3

### Epidemic Model with no Immunity

#### for $ \mathbf{R}_0 \gt 1$

```python
# code
Gamma = 0.22
Beta = 0.0004
h = 0.05

t = np.arange(0,10,h)
S = np.zeros(len(t))
I = np.zeros(len(t))

S[0] = 1945
I[0] = 315

def SM(t, S, I):
    return -1* Beta * I * S + Gamma * I

def IM(t, S, I):
    return Beta * I * S - Gamma * I

for i in range(0, len(t) -1):
    S[i+1] = S[i] + SM(t[i], S[i], I[i]) * h
    I[i+1] = I[i] + IM(t[i], S[i], I[i]) * h

# plt.plot(t, S, color='#FF00FF')
plt.plot(t, I, color='#0000FF')
plt.title('Population Model without Immunity')
# plt.legend(["Susceptible","Infective"],loc="lower right")
plt.legend(["Infective"],loc="lower right")
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```
