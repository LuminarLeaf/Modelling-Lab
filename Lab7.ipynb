```python
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1

```python
N = 101
x = np.zeros(N)
p = 0
q = 2
delx = (q-p)/(N-1)
x[0] = p
x[N-1] = q

for i in range(1,N-1):
    x[i] = x[i-1] + delx
```

```python
t = np.arange(0,10,2)
delt = 2
```

```python
u = np.zeros(N*len(t)).reshape(N,len(t))

for i in range(N):
    u[i][0] = 2 + x[i]*x[i]

for i in range(1,len(t)):
    u[N-1][i] = 6
```

```python
def rhs(x):
    return 0.05 - 1
```

```python
for i in range(0,N-1):
    for j in range(0,len(t)-1):
        u[i][j+1] = u[i][j] - u[i][j]*(delt/delx)*(u[i+1][j] - u[i][j]) + rhs(1)*delt
```

```python
plt.plot(x, u[:,3])
plt.show()
```
