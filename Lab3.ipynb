```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

## Kermack - McKendrick Model

```python
# code
t = np.arange(0,30,0.05)
S = np.zeros(len(t))
I = np.zeros(len(t))
R = np.zeros(len(t))

r = 0.0005
a = 0.2
h = 0.05

S[0] = 2000
I[0] = 300
R[0] = 0

def SM(t, S, I):
    return -1 * r * I * S

def IM(t, S, I):
    return r * I * S - a * I

def RM(t, I):
    return a * I

for i in range(0, len(t) -1):
    S[i+1] = S[i] + SM(t[i], S[i], I[i]) * h
    I[i+1] = I[i] + IM(t[i], S[i], I[i]) * h
    R[i+1] = R[i] + RM(t[i], I[i]) * h
```

### S,I,R vs t

```python
# figure 1
plt.plot(t, S, color='#FF00FF')
plt.plot(t, I, color='#0000FF')
plt.plot(t, R, color='#00FF00')
plt.legend(["Susceptible","Infective","Recovered"],loc="lower right")
plt.title('Population Model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

### S vs I

```python
# figure 2
plt.plot(I, S, color='#FF00FF')
plt.title('Phase Plane')
plt.xlabel('Susceptible')
plt.ylabel('Infective')
plt.show
```

## Model with Demography

```python
# code
t = np.arange(0,10,0.05)
S = np.zeros(len(t))
I = np.zeros(len(t))
R = np.zeros(len(t))

# a = 0.2
a = 2
# b = 1
b = 1/2300 
# R0 < 1
# r = 0.000005
# R0 > 1
r = 0.0005
h = 0.05

S[0] = 2000
I[0] = 300
R[0] = 0

def SM(t, S, I):
    return -1 * r * I * S - b * S + b

def IM(t, S, I):
    return r * I * S - a * I - b * I

def RM(t, R, I):
    return a * I - b * R

for i in range(0, len(t) -1):
    S[i+1] = S[i] + SM(t[i], S[i], I[i]) * h
    I[i+1] = I[i] + IM(t[i], S[i], I[i]) * h
    R[i+1] = R[i] + RM(t[i], R[i], I[i]) * h
```

### S,I,R vs t

```python
# figure 1
plt.plot(t, S, color='#FF00FF')
plt.plot(t, I, color='#0000FF')
plt.plot(t, R, color='#00FF00')
plt.legend(["Susceptible","Infective","Recovered"],loc="lower right")
plt.title('Population Model')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show
```

### S vs I

```python
# figure 2
plt.plot(I, S, color='#FF00FF')
plt.title('Phase Plane')
plt.xlabel('Susceptible')
plt.ylabel('Infective')
plt.show
```
