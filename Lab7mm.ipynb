## Problem 1

```python
# u(0,x) = 2+0.5*x*x

x1 = 0
xn = 2
nx = 101
hx = (xn-x1)/(nx-1)
x_ = []

i = x1
while i <= xn:
    x_.append(i)
    i += hx

t1 = 0
tn = 5
nt = 1001
ht = (tn-t1)/(nt-1)
t_ = []

i = t1
while i <= tn:
    t_.append(i)
    i += ht

u = []
```

```python
for i in t_:
    l = []
    for j in x_:
        l.append(0)
    u.append(l)
```

```python
for i in range(len(t_)):
    if i == 0:
        for j in range(len(x_)):
            u[i][j] = 2 + 0.5*(x_[j]**2)
    else:
        for j in range(len(x_)-1):
            u[i][j] = u[i-1][j] - u[i-1][j]*ht*(u[i-1][j+1] - u[i-1][j])/hx + (-0.95)*ht
        j = len(x_)-1
        u[i][j] = u[i-1][j] - u[i-1][j]*ht*(0 - u[i-1][j])/hx
```
