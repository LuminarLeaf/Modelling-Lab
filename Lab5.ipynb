# Problem 1

## Velocity profile of a Fluid particle

```python
# code

h = 0.05
t = np.arange(0, 10,h)
X = np.zeros(len(t))
Y = np.zeros(len(t))
Z = np.zeros(len(t))

X[0] = 1
Y[0] = 1
Z[0] = 1

def u(X, Y, t):
    return 2*X + 2*Y + 3*t

def v(X, Y, t):
    return X + Y + (t/2)

for i in range(0, len(t) -1):
    X[i+1] = X[i] + u(X[i], Y[i], t[i]) * h
    Y[i+1] = Y[i] + v(X[i], Y[i], t[i]) * h
    Z[i+1] = (X[i]**2 + Y[i]**2)**0.5
```

### X vs Y

```python
plt.plot(X, Y, color='#FF00FF')
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, max(max(X),max(Y)))
plt.ylim(0, max(max(X),max(Y)))
plt.show
```

### t vs Z

```python
plt.plot(t, Z, color='#FF00FF')
plt.title('t vs Z')
plt.xlabel('t')
plt.ylabel('Z')
plt.show
```

### t vs X

```python
plt.plot(t, X, color='#FF0000')
plt.title('t vs X')
plt.xlabel('t')
plt.ylabel('X')
plt.show
```

### t vs Y

```python
plt.plot(t, Y, color='#FF0000')
plt.title('t vs Y')
plt.xlabel('t')
plt.ylabel('Y')
plt.show
```

# Problem 2

## p 1

```python
# code
h = 0.05
X1 = np.arange(0, 10, h)
density = 1
u = np.zeros(len(X1))
u[0] = 0.5

def f_x_comp(x):
    return 0

def gaussian_div_p(x):
    return 3*x*x

def u_sol(u, x):
    return 2*(f_x_comp(x) - (1/density)*gaussian_div_p(x))

for i in range(0, len(X1) - 1):
    u[i+1] = u[i] + h * u_sol(u[i], X1[i])
```

```python
plt.plot(X1, u**0.5, color='red')
# plt.title('X vs Y')
plt.xlabel('displacement')
plt.ylabel('vel')
plt.show()
```

## p 2

```python
# code
h = 0.01
X2 = np.arange(-10, 0, h)
u1 = np.zeros(len(X2))
u1[0] = 0.25

def f_x1_comp(x):
    return -1*x

def gaussian_div_p1(x):
    return 2*x

def u1_sol(u, x):
    return 2*(f_x1_comp(x) - (1/density)*gaussian_div_p1(x))

for i in range(0, len(X2) - 1):
    u1[i+1] = u1[i] + h * u1_sol(u1[i], X2[i])
```

```python
plt.plot(X2, u1**0.5, color='#FF00FF')
# plt.title('X vs Y')
plt.xlabel('displacement')
plt.ylabel('vel')
plt.show
```
